# [dbo].[seleccionarRegistro]

```sql
/****** Object:  StoredProcedure [dbo].[seleccionarRegistro]    Script Date: 12/04/2025 10:40:31 p. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[seleccionarRegistro] (@nombreTabla VARCHAR(MAX), @nombreCampo VARCHAR(MAX), @valor VARCHAR(MAX))
AS
BEGIN
    SET NOCOUNT ON;

    DECLARE @sql_text NVARCHAR(MAX);

    -- Crear la consulta dinámica con los parámetros proporcionados
    SET @sql_text = N'SELECT * FROM ' + QUOTENAME(@nombreTabla) +
                    N' WHERE ' + QUOTENAME(@nombreCampo) + N' = @valor AND estatus = 1 ORDER BY id ASC;';

    -- Ejecutar la consulta dinámica usando sp_executesql y pasando el parámetro @valor
    EXEC sp_executesql @sql_text, N'@valor VARCHAR(MAX)', @valor;
END;

GO
```

```sql
/****** Object:  StoredProcedure [dbo].[seleccionarUsuario]    Script Date: 12/04/2025 10:40:31 p. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[seleccionarUsuario]
@correo VARCHAR(50),    
@pass VARCHAR(100)       
AS 
SELECT     
usuarios.id,     
usuarios.nombre,
usuarios.correo,   
usuarios.password
FROM       
usuarios
WHERE    
usuarios.correo = @correo   
AND usuarios.password = @pass
AND usuarios.estatus = 1    
GO
```
